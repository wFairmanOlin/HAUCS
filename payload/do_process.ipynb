```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
def get_df(file):
    df = pd.read_csv(file)
    df['pdo'] = df.do / 117.48 * 100
    return df
```

```python
df_baseline = get_df('data/2023-06-01-13-51-09.csv')
df_12_1 = get_df('data/2023-06-01-14-08-49.csv')
df_12_2 = get_df('data/2023-06-01-14-36-36.csv')

df_50_1 = get_df('data/2023-06-01-14-03-51.csv')
df_50_2 = get_df('data/2023-06-01-14-28-05.csv')

df_200_1 = get_df('data/2023-06-01-14-15-55.csv')
df_200_2 = get_df('data/2023-06-01-14-21-58.csv')

df_30_1 = get_df('data/2023-06-01-14-30-48.csv')

print(df_baseline.do.mean())

ref_time = np.array([0, 10, 20, 30, 40, 50, 60])
ref_12 = np.array([91, 16, 12, 11, 11, 11, 11])
ref_50 = np.array([102, 58, 57, 57, 56, 56, 56])
ref_50_1 = np.array([102, 58, 57, 57, 56, 56, 56])
# ref_200 = np.array([102, 199, 205, 208, 209, 212, 213])
ref_200 = np.array([103, 210, 222, 224, 222, 226, 227])
ref_200_1 = np.array([100, 238, 250, 251, 250, 251, 252])
```

```python
print("14 bit")
print("ysi: 11 probe: ", round(df_12_1.pdo[110:120].mean(), 3))
print("error: ", abs(11 - round(df_12_1.pdo[110:120].mean(), 3)))
print("10  bit")
print("ysi: 14 probe: ", round(df_12_2.pdo[110:120].mean(), 3))
print("error: ", abs(14 - round(df_12_2.pdo[110:120].mean(), 3)))

plt.figure()
plt.plot(df_12_1.time, df_12_1.pdo)
plt.plot(df_12_2.time, df_12_2.pdo)
plt.plot(ref_time, ref_12)
plt.xlabel("Time (secs)")
plt.ylabel("% Saturation")
plt.title("12% Saturation Tank")
plt.legend(['14 bit', '10 bit', 'ysi'])
plt.xlim(0, 60)
plt.show()
```
