```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df0 = pd.read_csv('data/2024-05-14-13-07-59.csv')
df1 = pd.read_csv('data/2024-05-14-13-53-48.csv')
df2 = pd.read_csv('data/2024-06-04-15-10-37.csv')
```

```python
print(df0.time[df0.temperature >= 37].iloc[0] / 60)
print(df1.time[df1.temperature >= 37].iloc[0] / 60)
print(df0.time[df0.temperature >= 36].iloc[0] / 60)
print(df1.time[df1.temperature >= 36].iloc[0] / 60)
print(df2.time[df2.temperature >= 36.5].iloc[5])
print(df2.time[df2.temperature >= 37.5].iloc[5])
```

```python
df0.head()

plt.figure(figsize=(11, 8))
plt.plot(df0.time/60, df0.temperature)
plt.plot(df1.time/60, df1.temperature)
plt.plot(df2.time/60, df2.temperature)
plt.hlines([27,38.5], 0, df0.time.max()/60, colors='r')
plt.xlabel('time (minutes)')
plt.ylabel('Temperature (C)')
plt.legend(['old housing', 'new housing', 'new pcb'], loc='best')
plt.savefig('data/temp_profile.png')
plt.show()
```
